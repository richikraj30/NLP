```python
import numpy as np
import pandas as pd
```

```python
reviews = pd.read_csv('all_kindle_review.csv')
```

```python
reviews
```

```python
reviews = reviews[['reviewText', 'rating']]
```

```python
reviews
```

```python
reviews['rating'] = reviews['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
reviews
```

```python
reviews['reviewText'] = reviews['reviewText'].str.lower()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from bs4 import BeautifulSoup

reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9]+', ' ', x))

# Removing stopwords
reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x:" ".join([word for word in x.split() if word not in stopwords.words('english')]))

## Remove url 
reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))

## Remove html tags
reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x: BeautifulSoup(x, 'html.parser').get_text())

## Remove any additional spaces
reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
reviews.shape
```

```python
corpus = reviews['reviewText']
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
words = [review.split() for review in corpus]
lemmatized_word = []
for review in words:
    lemmatized_review = [lemmatizer.lemmatize(word, pos="v") for word in review]
    lemmatized_word.append(lemmatized_review)
```

```python
corpus
```

```python
lemmatized_word
```

```python
import gensim
w2v_model = gensim.models.Word2Vec(lemmatized_word, epochs=1000, vector_size=100)
```

```python
w2v_model.wv.index_to_key
```

```python
w2v_model.corpus_count
```

```python
w2v_model.wv.similar_by_word('good')
```

```python
w2v_model.wv['good'].shape
```

```python
def avg_w2v(doc):
    vectors = [w2v_model.wv[word] for word in doc if word in w2v_model.wv.key_to_index]
    return np.mean(vectors, axis=0) if vectors else np.zeros(w2v_model.vector_size)
```

```python
X_w2v = [avg_w2v(review) for review in lemmatized_word]
X_new_w2v = np.array(X_w2v)
y = reviews['rating'].to_numpy()
```

```python
X_w2v
```

```python
df = pd.DataFrame(X_new_w2v)
df['Output'] = y
```

```python
X = df.drop(columns=['Output'])
y = df['Output']
```

```python
from sklearn.model_selection import train_test_split
X_train_w2v, X_test_w2v, y_train_w2v, y_test_w2v = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model_w2v = GaussianNB().fit(X_train_w2v, y_train_w2v)
```

```python
y_pred_w2v = nb_model_w2v.predict(X_test_w2v)
```

```python
from sklearn.metrics import accuracy_score
print("W2V accuracy: ", accuracy_score(y_test_w2v, y_pred_w2v))
```
