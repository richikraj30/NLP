### Preprocessing and data cleaning

```python
#  Considering the ratings below 3 as negative and above 3 as positive, so converting positive review as 1 and negative reviews as 0
reviews['rating'] = reviews['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
reviews
```

```python
reviews['rating'].value_counts()
```

```python
# Lowering all the cases
reviews['reviewText'] = reviews['reviewText'].str.lower()
```

```python
reviews.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
```

```python
# Removing special characters
reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9]+', ' ', x))

# Removing stopwords
reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x:" ".join([word for word in x.split() if word not in stopwords.words('english')]))

## Remove url 
reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))

## Remove html tags
reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x: BeautifulSoup(x, 'html.parser').get_text())

## Remove any additional spaces
reviews.loc[:, 'reviewText'] = reviews['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
reviews.head()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
reviews['reviewText'] = reviews['reviewText'].apply(lambda x: lemmatize_words(x))
```

```python
reviews.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(reviews['reviewText'], reviews['rating'], test_size = 0.20)
```

### Word embedding techniques

```python
X_train
```

```python
# BoW implemenatation
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer()
X_train_bow = bow.fit_transform(X_train).toarray()
X_test_bow = bow.transform(X_test).toarray()
```

```python
# TF-IDF implementation
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train).toarray()
X_test_tfidf = tfidf.transform(X_test).toarray()
```

```python
X_train_bow
```

```python
X_train_tfidf
```

### ML Algorithms

```python
from sklearn.naive_bayes import GaussianNB
nb_model_bow = GaussianNB().fit(X_train_bow, y_train)
nb_model_tfidf = GaussianNB().fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
y_pred_bow = nb_model_bow.predict(X_test_bow)
```

```python
y_pred_tfidf = nb_model_bow.predict(X_test_tfidf)
```

```python
confusion_matrix(y_test, y_pred_bow)
```

```python
print("BoW accuracy: ", accuracy_score(y_test, y_pred_bow))
```

```python
confusion_matrix(y_test, y_pred_tfidf)
```

```python
print("TFIDF accuracy: ", accuracy_score(y_test, y_pred_tfidf))
```

```python
import gensim
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
# words = []
# for sent in corpus:
#     sent_token = sent_tokenize(sent)
#     for sent in sent_token:
#         words.append(simple_preprocess(sent))

words = [review.split() for review in corpus]
```

```python
lemmatized_word = []
for review in words:
    lemmatized_review = [lemmatizer.lemmatize(word, pos="v") for word in review]
    lemmatized_word.append(lemmatized_review)
```

```python
lemmatized_word
```

```python
w2v_model = gensim.models.Word2Vec(lemmatized_word, epochs=1000, vector_size=100)
```

```python
w2v_model.wv.index_to_key
```

```python
w2v_model.corpus_count
```

```python
w2v_model.wv.similar_by_word('good')
```

```python
w2v_model.wv['good'].shape
```

```python
import numpy as np
def avg_w2v(doc):
    return np.mean([w2v_model.wv[word] for word in doc if word in w2v_model.wv.key_to_index], axis=0)
```

```python
from tqdm import tqdm
X_w2v = []
for i in tqdm(range(len(lemmatized_word))):
    X_w2v.append(avg_w2v(lemmatized_word[i]))
```
