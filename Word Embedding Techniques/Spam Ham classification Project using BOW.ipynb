```python
import pandas as pd
messages = pd.read_csv('spam.csv', sep=',', encoding='ISO-8859-1')
```

```python
messages
```

```python
messages = messages[['v1', 'v2']]
messages.columns = ['label', 'message']
```

```python
messages
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-z/^A-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Considering the values of Ham only
y = pd.get_dummies(messages['label'])
y = y.iloc[:, 0].values
```

```python
y
```

```python
# Train Test Spilt
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, ngram_range=(1,2))
```

## Creating Bag of Words

```python
#  independent features
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float = lambda x: "%.3g" % x))
X_train
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detection = MultinomialNB()
```

```python
# Trainig our data on X_tain and y_train
spam_detection.fit(X_train, y_train)
```

```python
# Storing the prediction fot X_test in y_pred for further analysis
y_pred = spam_detection.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
