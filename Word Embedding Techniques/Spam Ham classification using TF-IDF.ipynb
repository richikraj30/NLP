```python
import pandas as pd
messages = pd.read_csv('spam.csv', sep=',', encoding='ISO-8859-1')
```

```python
messages
```

```python
messages = messages[['v1', 'v2']]
messages.columns = ['label', 'message']
```

```python
import re
import nltk
```

```python
# Using lemmatizing instead of stemming
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
word_net = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-z/^A-Z]', ' ', messages['message'][i]).lower()
    review = review.split()
    review = [word_net.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
y = pd.get_dummies(messages['label'])
y = y.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
X_train = tfidf.fit_transform(X_train).toarray()
X_test = tfidf.transform(X_test).toarray()
```

## Creating TF-IDF

```python
from sklearn.naive_bayes import MultinomialNB
spam_detection = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred = spam_detection.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_test, y_pred)
```
